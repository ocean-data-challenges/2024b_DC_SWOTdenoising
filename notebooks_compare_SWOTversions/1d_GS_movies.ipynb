```python
import os
import glob
import sys
sys.path.append('..')
import numpy as np 
import xarray as xr 
import matplotlib.pyplot as plt
from datetime import datetime
import glob 
#import warnings
#warnings.filterwarnings('ignore')

from src import mod_plot
from src import mod_switchvar
from src.dict_regions_info import *
```

```python

region_info = RegionInfo('GS_info').region_info   # Region dictionary info from dict_regions_info.py
```

<div class="alert alert-info" role="alert">

<h2>1. Read from regional data file </h2>

</div>
 

```python
path_data = '../swot_versions/SWOT_v1.0/SWOT_'+region_info['name']+'_v1.0.nc'

starting_date = np.datetime64('2023-04-23')

label_noisy = 'Noisy'   # Name for plots title and saving figures
label_denoised = 'Denoised'   # Name for plots title and saving figures


name_noisy_v1 = 'ssha' # Name of the denoised SWOT variable in the xarray file 
name_denoised_v1 = 'ssha_noiseless' # Name of the denoised SWOT variable in the xarray file

orbit = '1d'

ds_passes_v1 = xr.open_mfdataset(path_data,combine='nested',concat_dim='num_lines')
ds_passes_v1
```

```python
path_data = '../swot_versions/SWOT_v2.0/SWOT_'+region_info['name']+'_v2.0.nc'

starting_date = np.datetime64('2023-04-23')

label_noisy = 'Noisy'   # Name for plots title and saving figures
label_denoised = 'Denoised'   # Name for plots title and saving figures


name_noisy_v2 = 'ssha_unfiltered' # Name of the denoised SWOT variable in the xarray file 
name_denoised_v2 = 'ssha_filtered' # Name of the denoised SWOT variable in the xarray file

orbit = '1d'

ds_passes_v2 = xr.open_mfdataset(path_data,combine='nested',concat_dim='num_lines')
ds_passes_v2
```

<div class="alert alert-info" role="alert">

<h2>2. Create SSH movie </h2>

</div> 

```python
mod_plot.movie_intercomp(ds_passes_v1, methods=[label_noisy,label_denoised], var_type='ssh', name_var=[name_noisy_v1,name_denoised_v1],date=starting_date, dir_output='../figures/',
                    region_info=region_info, vmax=0.5, suffix='v1', framerate=3, colsize = 14)
```

```python
mod_plot.movie_intercomp(ds_passes_v2, methods=[label_noisy,label_denoised], var_type='ssh', name_var=[name_noisy_v2,name_denoised_v2],date=starting_date, dir_output='../figures/',
                    region_info=region_info, vmax=0.5, suffix='v2', framerate=3, colsize = 14)
```

<div class="alert alert-info" role="alert">

<h2>3. Compute SSH gradients and Laplacian </h2>

</div>
 

```python
grad_noisy, lapl_noisy = mod_switchvar.deriv1and2(ds_passes_v1[name_noisy])
grad_denoised, lapl_denoised = mod_switchvar.deriv1and2(ds_passes_v1[name_denoised])   

ds_passes_v1 = ds_passes_v1.assign(grad_noisy=lambda ds_passes_v1: ds_passes_v1[name_noisy] * 0 + grad_noisy) 
ds_passes_v1 = ds_passes_v1.assign(grad_denoised=lambda ds_passes_v1: ds_passes_v1[name_denoised] * 0 + grad_denoised) 

ds_passes_v1 = ds_passes_v1.assign(lapl_noisy=lambda ds_passes_v1: ds_passes_v1[name_noisy] * 0 + lapl_noisy) 
ds_passes_v1 = ds_passes_v1.assign(lapl_denoised=lambda ds_passes_v1: ds_passes_v1[name_denoised] * 0 + lapl_denoised) 
```

```python
grad_noisy, lapl_noisy = mod_switchvar.deriv1and2(ds_passes_v2[name_noisy])
grad_denoised, lapl_denoised = mod_switchvar.deriv1and2(ds_passes_v2[name_denoised])   

ds_passes_v2 = ds_passes_v2.assign(grad_noisy=lambda ds_passes_v2: ds_passes_v2[name_noisy] * 0 + grad_noisy) 
ds_passes_v2 = ds_passes_v2.assign(grad_denoised=lambda ds_passes_v2: ds_passes_v2[name_denoised] * 0 + grad_denoised) 

ds_passes_v2 = ds_passes_v2.assign(lapl_noisy=lambda ds_passes_v2: ds_passes_v2[name_noisy] * 0 + lapl_noisy) 
ds_passes_v2 = ds_passes_v2.assign(lapl_denoised=lambda ds_passes_v2: ds_passes_v2[name_denoised] * 0 + lapl_denoised) 
```
